 **Importing the Dependencies**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

**Data Collection & Processing**



```python
#loading the data
data=sklearn.datasets.load_breast_cancer()
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
df.head(5)
```

```python
df.shape
```

```python
df['target']=data.target
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
df['target'].value_counts()
```

```python
df.groupby('target').mean()
```

```python
x=df.drop(columns='target',axis=1)
y=df['target']
```

**Spliting the data into training and testing**

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
train_x=scalar.fit_transform(train_x)
test_x=scalar.transform(test_x)
```

```python
train_x.shape
```

**Building Neural Network**

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)), #flatten is converting into single dimension array
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(2,activation='sigmoid')#here ,we have only 2 classes
    ])
```

```python
#compiling the neural network

model.compile(

              optimizer='adam',
              loss='sparse_categorical_crossentropy',#for categorical vaue only
              metrics=['accuracy']
)
```

```python
history=model.fit(train_x,train_y,validation_split=0.1,epochs=10)
```

```python
plt.plot(history.history['val_accuracy'])
plt.plot(history.history['accuracy'])


plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training data','validation_data'],loc='lower right')
```

```python
plt.plot(history.history['val_loss'])
plt.plot(history.history['loss'])


plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training data','validation_data'],loc='upper right')
```

```python
loss,accuracy=model.evaluate(test_x,test_y)
print("Accuracy:",accuracy)
```
